```python
# Import the packages needed for this project
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv('/Volumes/Vandy Main/Portfolio Projects/Correlation in Python/movies.csv')
```

```python
# view data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# There are many ways to handle missing data, but for now we will drop all rows that have missing data

df = df.dropna()
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change the data type in the budget and gross columns from float64 to int64

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# The year column does not always match the year in the released column.
# We will create a correct year column using the year in the released column.

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
# Sort data in descending order by gross revenue 

df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Drop duplicates (if there are any)

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget per Film')
plt.show()
```

```python
# Regression plot using seaborn.  Using budget to predict gross

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```

```python
# Obtain a correlation matrix with the numerical fields
# Correlation between budget and gross = .74
# r-squared = .5476. This means that 54.76% of the variation in gross revenue can be predicted by budget

df.corr()
```

```python
# Regression plot using budget to predict score
# Correlation between budget and votes = .44
# r-squared = .1936. This means that 19.36% of the variation in votes can be predicted by budget

sns.regplot(x = 'budget', y = 'votes', data = df, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```
